# Import Libraries

```python
# Import Libaries
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

# Functions

```python
# Function to one hot encode selected columns  
def hot_encode_cat_col(dataframe,list_cat_col):
    for col in list_cat_col:    
        dataframe = pd.get_dummies(columns=[col],drop_first=True,data=dataframe) 
    return dataframe
```

```python
# Function to loop through the column list and drops the columns
def drop_columns(dataframe,col_list):
    for col in col_list:
        dataframe.drop(columns=[col], inplace=True)
           
```

# Load Dataset

```python
# Load in Dataset
combined_resale_df = pd.read_csv('../assets/data_clean/resale_combine_clean.csv')
```

# One Hot Encode Object Column

```python
# view dataframe info()
combined_resale_df.info(verbose=True)
```

```python
# Get columns with datatype object
string_columns = combined_resale_df.select_dtypes(include=['object']).columns.tolist()
string_columns
```

```python
# View the number of unique value for each column value to decide what to encode
# combined_resale_df['town'].unique()

for string_col in string_columns:
    print(string_col,": ", len(combined_resale_df[string_col].unique()))
```

```python
# view first 2 rows of data
combined_resale_df.head(2)
```

- Observed that block, street name and address has high unique values.
    - These columns will be dropped as location is covered by "planning_area"
- mth_format will be drop as there is already year and month column

```python
# Drop columns
col_to_drop = ['block', 'street_name', 'address',  'month_fmt']
drop_columns(combined_resale_df,col_to_drop)
```

```python
# Check that the columns has been dropped and update string_columns
string_columns = combined_resale_df.select_dtypes(include=['object']).columns.tolist()
string_columns
```

- Observed that the columns has been dropped
- Next: proceed with one hot encoding for remaining columns

```python
# create a copy of dataframe for one hot encoding
combined_resale_df_copy = combined_resale_df.copy()

# One Hot Encode all String Columns
combined_resale_df_copy = hot_encode_cat_col(combined_resale_df_copy,string_columns)
```

```python
# Check that there are no more string columns
combined_resale_df_copy.select_dtypes(include=['object']).columns.tolist()
```

- Validated that there are no more string columns where dtype is object
